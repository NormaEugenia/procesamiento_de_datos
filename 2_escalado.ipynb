# Escalado
Conjunto de datos: iris (incluido en sklearn)
Este conjunto de datos es compacto y perfecto para demostrar técnicas de escalado.

```python
# Importar librerías necesarias
from sklearn.datasets import load_iris
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Cargar el conjunto de datos
iris = load_iris(as_frame=True)
data = iris.data
```

```python
# Visualizar las estadísticas iniciales
print("Estadísticas iniciales:")
data.describe()
```

```python
# Aplicar MinMaxScaler
minmax_scaler = MinMaxScaler()
data_minmax = pd.DataFrame(minmax_scaler.fit_transform(data), columns=data.columns)
```

```python
# Aplicar StandardScaler
standard_scaler = StandardScaler()
data_standard = pd.DataFrame(standard_scaler.fit_transform(data), columns=data.columns)
```

### 1. ¿Qué diferencias encuentras entre las estadísticas originales y las del escalado MinMax?


```python
Al aplicar MinMaxScaler,
las estadísticas de los datos cambian porque cada variable se reescala para que sus valores queden dentro del rango de
 0 a 1. Esto hace que el valor mínimo de cada columna sea 0 y el máximo sea 1, mientras que la media, los cuartiles
 y la desviación estándar se ajustan proporcionalmente a esta nueva escala. Sin embargo, el escalado no modifica la forma
  de la distribución ni el orden relativo de los datos, por lo que las relaciones entre las observaciones se conservan.
   En comparación con los datos originales, las variables dejan de tener sus unidades y rangos originales,
   pero se vuelven más comparables entre sí, lo que puede mejorar el desempeño de algunos algoritmos de aprendizaje automático.
```

### 2. Crea una gráfica de la distribución de la variable 'sepal length (cm)' antes y después del escalado.

```python
ris = load_iris(as_frame=True)
data = iris.data
minmax_scaler = MinMaxScaler()
data_minmax = pd.DataFrame(
    minmax_scaler.fit_transform(data),
    columns=data.columns
)


plt.figure(figsize=(8, 5))

plt.hist(
    data['sepal length (cm)'],
    bins=15,
    alpha=0.6,
    label='Original'
)

plt.hist(
    data_minmax['sepal length (cm)'],
    bins=15,
    alpha=0.6,
    label='Escalado MinMax'
)

plt.title('Distribución de Sepal Length antes y después del escalado')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.legend()
plt.grid(True, alpha=0.3)

plt.show()
```

### 3. Crea una gráfica de dispersión entre 'sepal length (cm)' y 'sepal width (cm)' con los datos escalados por StandardScaler.

```python
standard_scaler = StandardScaler()
data_standard = pd.DataFrame(
    standard_scaler.fit_transform(data),
    columns=data.columns
)

# Gráfica de dispersión
plt.figure(figsize=(8, 6))

plt.scatter(
    data_standard['sepal length (cm)'],
    data_standard['sepal width (cm)'],
    alpha=0.7
)

plt.title('Sepal Length vs Sepal Width (Datos Escalados con StandardScaler)')
plt.xlabel('Sepal Length (cm) Escalado')
plt.ylabel('Sepal Width (cm) Escalado')
plt.grid(True, alpha=0.3)

plt.show()
```

### 4. ¿Qué técnica de escalado crees que es más adecuada para este conjunto de datos? ¿Por qué?

```python
Para el conjunto de datos Iris, considero que StandardScaler
 suele ser la técnica más adecuada porque las variables tienen distribuciones relativamente
  normales y no están limitadas naturalmente a un rango específico. Este método centra los datos alrededor de una
   media de 0 y una desviación estándar de 1, lo que facilita que muchos algoritmos de aprendizaje automático, como la
    regresión logística, las máquinas de vectores de soporte (SVM) y el análisis de componentes principales (PCA), funcionen de manera más eficiente. Aunque
    MinMaxScaler
     también es una buena opción y conserva los datos en un rango de 0 a 1, puede verse más afectado
     por valores atípicos. Como el conjunto Iris no presenta outliers extremos y es común utilizarlo con algoritmos sensibles
     a la escala y a la varianza de las características, **StandardScaler generalmente ofrece una mejor estandarización de los datos sin alterar
     las relaciones entre las observaciones.
```
