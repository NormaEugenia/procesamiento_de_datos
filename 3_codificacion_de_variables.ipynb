### 1. ¿Qué diferencias encuentras entre LabelEncoder y OneHotEncoder?

```python
La principal diferencia entre LabelEncoder y OneHotEncoder
es la forma en que representan las variables categóricas. LabelEncoder asigna un número entero a
cada categoría (por ejemplo, "rojo" = 0, "azul" = 1, "verde" = 2), lo que es sencillo y ocupa menos espacio,
 pero puede hacer que algunos algoritmos interpreten erróneamente que existe un orden o jerarquía entre las categorías. En cambio, OneHotEncoder
  crea una nueva columna para cada categoría y utiliza valores binarios (0 o 1) para indicar su presencia, evitando cualquier interpretación de orden entre ellas.
  Por esta razón, OneHotEncoder suele ser más adecuado para variables categóricas nominales, mientras que LabelEncoder se utiliza principalmente
   para codificar etiquetas de salida o categorías que realmente poseen un orden natural.
```

### 4. Si quisieras aplicar un modelo de aprendizaje automático, ¿qué tipo de codificación elegirías para las variables categóricas? Explica tu respuesta.

```python
Para aplicar un modelo de aprendizaje automático, en la mayoría de los casos elegiría OneHotEncoder para variables
categóricas nominales (sin orden), porque evita que el modelo interprete relaciones falsas de orden entre categorías.
Por ejemplo, si una variable tiene valores como “rojo”, “azul” y “verde”, codificarlas como 0, 1 y 2 con LabelEncoder puede inducir
a algoritmos como regresión logística, SVM o redes neuronales a pensar que existe una jerarquía o distancia entre ellas,
cuando en realidad no la hay. En cambio, OneHotEncoder crea una representación binaria independiente para cada categoría,
lo que permite que el modelo aprenda su efecto sin asumir orden.

Sin embargo, usaría **LabelEncoder solo en casos específicos, como cuando
 la variable objetivo (y) es categórica en un problema de clasificación, o cuando se trata de variables
 ordinales (por ejemplo, “bajo”, “medio”, “alto”), donde sí existe un orden natural. También puede ser útil
 en modelos basados en árboles (como Random Forest o XGBoost), ya que estos son menos sensibles al orden artificial que crea la codificación numérica.
```
